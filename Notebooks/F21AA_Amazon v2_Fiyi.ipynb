# **Text Processing and Normalization**

```
# Define text preprocessing functions
def basic_preprocessing(text):
    """Basic preprocessing: lowercase and remove special characters"""
    text = str(text).lower()  # Convert to string first in case of non-string values
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    return text
```

```
def remove_stopwords(text):
    """Remove stopwords"""
    stop_words = set(stopwords.words('english'))
    tokens = word_tokenize(text)
    tokens = [word for word in tokens if word not in stop_words]
    return ' '.join(tokens)
```

```
def apply_stemming(text):
    """Apply Porter stemming"""
    stemmer = PorterStemmer()
    tokens = word_tokenize(text)
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    return ' '.join(stemmed_tokens)
```

```
def apply_lemmatization(text):
    """Apply lemmatization"""
    lemmatizer = WordNetLemmatizer()
    tokens = word_tokenize(text)
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    return ' '.join(lemmatized_tokens)
```

```
# Function to apply full preprocessing pipeline (we'll use different combinations later)
def preprocess_text(text, remove_stop=True, stem=False, lemmatize=False):
    """Full preprocessing pipeline with options"""
    text = basic_preprocessing(text)
    if remove_stop:
        text = remove_stopwords(text)
    if stem:
        text = apply_stemming(text)
    elif lemmatize:
        text = apply_lemmatization(text)
    return text
```

# **Model Text as a Sequence**

```
# We'll use the original text with basic preprocessing
X_train_seq, X_val_seq, y_train_seq, y_val_seq = train_test_split(
    train_df['Text'].astype(str).apply(basic_preprocessing),
    train_df['Score'],
    test_size=0.2,
    random_state=42,
    stratify=train_df['Score']
)
```

```
# Convert classes to 0-based indexing for neural networks
y_train_seq = y_train_seq - 1
y_val_seq = y_val_seq - 1

# Tokenize the text
max_words = 10000
max_sequence_length = 200

tokenizer = Tokenizer(num_words=max_words)
tokenizer.fit_on_texts(X_train_seq)

X_train_seq = tokenizer.texts_to_sequences(X_train_seq)
X_val_seq = tokenizer.texts_to_sequences(X_val_seq)
```

```
# Pad sequences to ensure uniform length
X_train_seq = pad_sequences(X_train_seq, maxlen=max_sequence_length)
X_val_seq = pad_sequences(X_val_seq, maxlen=max_sequence_length)
```

```
# 5.1 Simple RNN Model
def build_simple_rnn():
    model = Sequential()
    model.add(Embedding(max_words, 100, input_length=max_sequence_length))
    model.add(LSTM(128, dropout=0.2, recurrent_dropout=0.2))
    model.add(Dense(5, activation='softmax'))  # 5 classes (ratings 1-5)

    model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    return model
```

```
# Train the RNN model
rnn_model = build_simple_rnn()
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

rnn_history = rnn_model.fit(
    X_train_seq, y_train_seq,
    epochs=10,
    batch_size=64,
    validation_data=(X_val_seq, y_val_seq),
    callbacks=[early_stopping]
)
```

```
# Evaluate the RNN model
rnn_val_loss, rnn_val_acc = rnn_model.evaluate(X_val_seq, y_val_seq)
print(f"RNN Validation Accuracy: {rnn_val_acc:.4f}")
```

```
# 5.2 Bidirectional LSTM with GloVe embeddings
# First, download and prepare GloVe embeddings
!wget -q -O glove.6B.100d.txt.zip https://nlp.stanford.edu/data/glove.6B.zip
!unzip -q glove.6B.100d.txt.zip
```

```
# Load GloVe embeddings
embedding_dim = 100
embedding_index = {}
with open('glove.6B.100d.txt', encoding='utf-8') as f:
    for line in f:
        values = line.split()
        word = values[0]
        coefs = np.asarray(values[1:], dtype='float32')
        embedding_index[word] = coefs
```

```
# Prepare embedding matrix
embedding_matrix = np.zeros((max_words, embedding_dim))
for word, i in tokenizer.word_index.items():
    if i < max_words:
        embedding_vector = embedding_index.get(word)
        if embedding_vector is not None:
            embedding_matrix[i] = embedding_vector
```

```
# Build the BiLSTM model with GloVe embeddings
def build_bilstm_glove():
    model = Sequential()
    model.add(Embedding(max_words,
                        embedding_dim,
                        weights=[embedding_matrix],
                        input_length=max_sequence_length,
                        trainable=False))
    model.add(Bidirectional(LSTM(128, dropout=0.2, recurrent_dropout=0.2)))
    model.add(Dense(64, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(5, activation='softmax'))

    model.compile(loss='sparse_categorical_crossentropy',
                  optimizer='adam',
                  metrics=['accuracy'])
    return model
```

```
# Train the BiLSTM model
bilstm_model = build_bilstm_glove()
early_stopping = EarlyStopping(monitor='val_loss', patience=3, restore_best_weights=True)

bilstm_history = bilstm_model.fit(
    X_train_seq, y_train_seq,
    epochs=10,
    batch_size=64,
    validation_data=(X_val_seq, y_val_seq),
    callbacks=[early_stopping]
)
```

```
# Evaluate the BiLSTM model
bilstm_val_loss, bilstm_val_acc = bilstm_model.evaluate(X_val_seq, y_val_seq)
print(f"BiLSTM with GloVe Validation Accuracy: {bilstm_val_acc:.4f}")
```

# **Topic Modeling of High and Low Ratings**

```
# Prepare data for topic modeling
# Get reviews with rating 1 (low) and 5 (high)
low_rating_reviews = train_df[train_df['Score'] == 1]['Text'].astype(str).apply(lambda x: preprocess_text(x, remove_stop=True))  # Changed both column names
high_rating_reviews = train_df[train_df['Score'] == 5]['Text'].astype(str).apply(lambda x: preprocess_text(x, remove_stop=True))  # Changed both column names
```

```
if len(low_rating_reviews) > 5000:
    low_rating_reviews = low_rating_reviews.sample(5000, random_state=42)
if len(high_rating_reviews) > 5000:
    high_rating_reviews = high_rating_reviews.sample(5000, random_state=42)
```

```
# Process text for topic modeling
def prepare_for_lda(texts):
    # Tokenize
    tokenized_texts = [text.split() for text in texts]

    # Create dictionary
    dictionary = Dictionary(tokenized_texts)

    # Filter extremes (optional)
    dictionary.filter_extremes(no_below=5, no_above=0.5)

    # Convert to bag of words
    corpus = [dictionary.doc2bow(text) for text in tokenized_texts]

    return dictionary, corpus
```

```
# Topic modeling for low ratings
low_dictionary, low_corpus = prepare_for_lda(low_rating_reviews)
low_lda_model = LdaModel(
    corpus=low_corpus,
    id2word=low_dictionary,
    num_topics=10,
    random_state=42,
    passes=10
)
```

```
# Topic modeling for high ratings
high_dictionary, high_corpus = prepare_for_lda(high_rating_reviews)
high_lda_model = LdaModel(
    corpus=high_corpus,
    id2word=high_dictionary,
    num_topics=10,
    random_state=42,
    passes=10
)
```

```
# Print the topics
print("Topics in Low Rating Reviews:")
for i, topic in low_lda_model.print_topics(num_words=10):
    print(f"Topic {i}: {topic}")

print("\nTopics in High Rating Reviews:")
for i, topic in high_lda_model.print_topics(num_words=10):
    print(f"Topic {i}: {topic}")

def visualize_topics(lda_model, title):
    """
    Visualize top words for each topic with a horizontal bar chart design
    that completely prevents label overlap.
    """
    n_topics = lda_model.num_topics
    n_words = 8  # Reduce number of words per topic to reduce clutter
    
    # Create one separate figure for each topic to avoid overlap
    for topic_idx in range(n_topics):
        # Get top words and weights for this topic
        topic_words = dict(lda_model.show_topic(topic_idx, topn=n_words))
        words = list(topic_words.keys())
        weights = list(topic_words.values())
        
        # Create a figure for this topic
        plt.figure(figsize=(10, 6))
        
        # Create horizontal bars with more spacing
        y_pos = np.arange(len(words)) * 1.2  # Add extra spacing between bars
        plt.barh(y_pos, weights, height=0.8, color='steelblue')
        
        # Set y-tick labels and positions
        plt.yticks(y_pos, words, fontsize=12)
        
        # Add title and labels
        plt.title(f'{title} - Topic {topic_idx}', fontsize=14)
        plt.xlabel('Weight', fontsize=12)
        
        # Remove y-axis label
        plt.ylabel('')
        
        # Adjust layout and display
        plt.tight_layout()
        plt.show()
```

```
visualize_topics(low_lda_model, "Topics in Low Rating Reviews")
visualize_topics(high_lda_model, "Topics in High Rating Reviews")
```
